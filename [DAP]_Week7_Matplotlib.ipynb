## Kiến trúc matplotlib

Gồm 3 tầng

- backend layer
- artist layer
- scripting layer

**Backend Layer**: là lớp dưới cùng của hình, bao gồm việc thực hiện các chức năng khác nhau cần thiết cho việc vẽ biểu đồ. Có ba lớp thiết yếu:

- FigureCanvas: Bề mặt mà hình sẽ được vẽ
- Renderer: Lớp chăm sóc bản vẽ trên bề mặt
- Event: Nó xử lý các sự kiện chuột và bàn phím

**Artist Layer**: Là tầng thứ 2, nó chịu trách nhiệm về các chức năng vẽ khác nhau, như trục, điều phối cách sử dụng trình kết xuất trên FigureCanvas.

**Scripting Layer**: là lớp trên cùng mà hầu hết các mã code của chúng ta sẽ chạy trên đó. Các hàm trong lớp script, gần như tự động đảm nhiệm các lớp khác, và tất cả những gì chúng ta cần quan tâm là trạng thái hiện tại (figure & subplot).

## Các khái niệm trong matplotlib

**Hình vẽ (Figure)**: Đây là một hình vẽ hoàn chỉnh có thể chứa một hoặc nhiều trục - axes (đồ thị - plots). Chúng ta có thể xem Figure như một bảng vẽ (canvas) chứa các plots.

**Trục (Axes)**: Một Figure có thể chứa nhiều Trục. Nó bao gồm hai hoặc ba đối tượng Trục (trong trường hợp 3D). Mỗi Trục bao gồm một title, một nhãn x và một nhãn y.

**Trục (Axis)**: Trục là các đối tượng giống như dòng và chịu trách nhiệm cho việc tạo ra giới hạn của đồ thị.

**Nghệ sĩ (Artist)**: Một nghệ sĩ là tất cả những gì chúng ta thấy trên đồ thị như các đối tượng Văn bản (Text), đối tượng Line2D và đối tượng collection. Hầu hết các Artist đều liên kết với Trục.

## Axes class

> **`ax=fig.add_axes([0,0,1,1])`**

- Đối tượng Axes là khu vực của hình ảnh trong không gian dữ liệu. Một figure cụ thể có thể chứa nhiều đối tượng Axes, nhưng một đối tượng Axes cụ thể chỉ có thể nằm trong một Hình vẽ (Figure). Axes chứa hai (hoặc ba trong trường hợp 3D) đối tượng Axis (Axis). Class Axes và các functions của nó là thành phần cơ bản để làm việc với giao diện hướng đối tượng.

- Đối tượng Axes được thêm vào hình vẽ bằng cách gọi phương thức add_axes(). Nó trả về đối tượng Axes và thêm một Axes tại vị trí rect [left, bottom, width, height] trong đó tất cả các đại lượng đều được tính bằng phần trăm chiều rộng và chiều cao của hình vẽ.




### Legend method

**`ax.legend(handles, labels, loc)`**

- Phương thức legend() của lớp Axes thêm một chú thích vào hình vẽ đồ thị. Nó nhận ba tham số.

- Trong đó, labels là một mảng các strings và handles là một mảng các đối tượng Line2D hoặc Patch. loc có thể là một string hoặc một số integer chỉ định vị trí của legend.



image.png

````python
y = [1, 4, 9, 16, 25,36,49, 64]
x1 = [1, 16, 30, 42,55, 68, 77,88]
x2 = [1,6,12,18,28, 40, 52, 65]

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(x1, y, 'ys-') # solid line with yellow colour and square marker
ax.plot(x2, y, 'go--') # dash line with green colour and circle marker

ax.legend(labels = ('tv', 'Smartphone'), loc = 'lower right') # legend placed at lower right
ax.set_title("Advertisement effect on sales")
ax.set_xlabel('medium')
ax.set_ylabel('sales')

# plt.legend(labels=('tv', 'Smartphone'), loc='lower right')
# plt.title("Advertisement effect on sales")
# plt.xlabel('medium')
# plt.ylabel('sales')
plt.show()
````

## Multiplots



### plt.subplot()

> **`plt.subplot(nrows, ncols, index)`**


Hàm subplot() trả về đối tượng Axes ở vị trí grid cho trước.

- Trong hình vẽ hiện tại, hàm này tạo và trả về một đối tượng Axes tại vị trí index của một grid có nrows hàng và ncols cột. Các chỉ mục đi từ 1 đến nrows * ncols, tăng dần theo thứ tự hàng. Nếu nrows, ncols và chỉ mục đều nhỏ hơn 10, các chỉ mục cũng có thể được đưa ra dưới dạng một số có ba chữ số được nối lại.

- Ví dụ, subplot(2, 3, 3) và subplot(233) đều tạo ra một đối tượng Axes ở góc phải trên của hình vẽ hiện tại, chiếm một nửa chiều cao của hình vẽ và một phần ba chiều rộng của hình vẽ.

- Việc tạo một subplot sẽ xóa bất kỳ subplot trước đó nào trùng lấn với nó ngoài việc chia sẻ một ranh giới.

````python
## You can add an insert plot in the same figure by adding another axes object in the same figure canvas.

x = np.arange(0, np.pi*2, 0.05)
y = np.sin(x)

fig=plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # main axes
axes2 = fig.add_axes([0.55, 0.55, 0.3, 0.3]) # inset axes

axes1.plot(x, y, 'b')
axes2.plot(x, np.cos(x), 'r')

axes1.set_title('sine')
axes2.set_title("cosine")

plt.show()
````

````python
fig, a =  plt.subplots(2,2)
x = np.arange(1,5)

a[0][0].plot(x, x*x)
a[0][0].set_title('square')

a[0][1].plot(x, np.sqrt(x))
a[0][1].set_title('square root')

a[1][0].plot(x, np.exp(x))
a[1][0].set_title('exp')

a[1][1].plot(x, np.log10(x))
a[1][1].set_title('log')

plt.show()
````

### plt.subplot2grid()

````python
a1 = plt.subplot2grid((3,3),(0,0),colspan = 2)
a2 = plt.subplot2grid((3,3),(0,2), rowspan = 3)
a3 = plt.subplot2grid((3,3),(1,0),rowspan = 2, colspan = 2)
x = np.arange(1,10)

a2.plot(x, x*x)
a2.set_title('square')

a1.plot(x, np.exp(x))
a1.set_title('exp')

a3.plot(x, np.log(x))
a3.set_title('log')

plt.tight_layout()
plt.show()
````

# Box plot ( Whisker plot)


- Biểu đồ box là một cách tốt để hiển thị nhiều đặc điểm quan trọng của dữ liệu định lượng (số liệu).

- Nó hiển thị trung vị của dữ liệu. Đây là giá trị ở giữa của dữ liệu và là một loại giá trị trung bình.

- Nó cũng hiển thị phạm vi và các phân vị của dữ liệu. Điều này cho chúng ta biết một cái gì đó về cách dữ liệu phân tán.

- Dưới đây là biểu đồ hộp của tuổi của tất cả các người đoạt giải Nobel đến năm 2020:



image.png


https://www.w3schools.com/statistics/statistics_box_plots.php


- **Trung vị (median)** là đường màu đỏ đi qua giữa của 'hộp'. Chúng ta có thể thấy rằng điều này chỉ hơi trên con số 60 trên số dưới dòng. Vì vậy, giá trị trung bình của tuổi là 60 tuổi.

- Phía trái của hộp là **phân vị thứ nhất (1st quartile)**. Đây là giá trị phân chia ra khỏi tứ thứ nhất, hoặc **25%** của dữ liệu, so với phần còn lại. Ở đây, đó là 51 tuổi.

- Phía phải của hộp là **phân vị thứ ba (3rd quartile)**. Đây là giá trị phân chia ra khỏi ba tứ thứ nhất, hoặc **75%** của dữ liệu, so với phần còn lại. Ở đây, đó là 69 tuổi.

- Khoảng cách giữa hai bên của hộp được gọi là **phạm vi liên phân vị (inter-quartile range - IQR)**. Điều này cho chúng ta biết vị trí của **'nửa giữa'** các giá trị. Ở đây, một nửa các người chiến thắng có tuổi từ 51 đến 69 tuổi.

- Các đầu của các đường từ hộp ở bên trái và bên phải là giá trị tối thiểu và tối đa trong dữ liệu. Khoảng cách giữa chúng được gọi là phạm vi.

- Người chiến thắng trẻ nhất là 17 tuổi và người lớn tuổi nhất là 97 tuổi. Vì vậy, phạm vi của tuổi của các người chiến thắng là 80 tuổi.


Thư viện matplotlib cung cấp hàm boxplot() trong module matplotlib.pyplot, giúp chúng ta tạo biểu đồ hộp.

```
pyplot.boxplot(data, notch=None, vert=None, patch_artist=None, widths=None, .....)
```



*   notch = True attribute creates the notch format to the box plot
*   patch_artist = True fills the boxplot with colors
*   vert = 0 attribute creates horizontal box plot



## Simple example

````python
# Creating dataset
np.random.seed(10)
data = np.random.normal(100, 20, 200)

# Creating plot
plt.boxplot(data)

# show plot
plt.show()
````

````python
plt.boxplot(data, notch=True)
plt.show()
````

````python
plt.boxplot(data, notch=True, patch_artist = True)
plt.show()
````

````python
plt.boxplot(data, notch=True, patch_artist = True, vert = 0)
plt.show()
````

````python
box = plt.boxplot(data, notch=True, patch_artist=True)
keys = box.keys()
print(keys)

box["boxes"][0].set(color="red")
box["boxes"][0].set(facecolor="green")

box["fliers"][0].set(markeredgecolor="blue")

box["whiskers"][0].set(color="orange")
box["whiskers"][1].set(color="orange")

box["caps"][0].set(color="yellow")
box["caps"][1].set(color="yellow")

plt.show()
````

````python
import numpy as np
np.random.seed(10)

data1 = np.random.normal(100, 10, 200)
data2 = np.random.normal(80, 30, 200)
data3 = np.random.normal(90, 20, 200)
data4 = np.random.normal(70, 25, 200)
data_to_plot = [data1, data2, data3, data4]
````

````python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

bp = ax.boxplot(data_to_plot)
plt.show()
````

````python
fig = plt.figure(figsize =(10, 7))
ax = fig.add_subplot(111)

# Creating axes instance
bp = ax.boxplot(data_to_plot, patch_artist = True, notch ='True')

colors = ['#0000FF', '#00FF00', '#FFFF00', '#FF00FF']

for patch, color in zip(bp['boxes'], colors):
    patch.set_facecolor(color)

# changing color and linewidth of whiskers
for whisker in bp['whiskers']:
    whisker.set(color ='#8B008B', linewidth = 1.5, linestyle =":")

# changing color and linewidth of caps
for cap in bp['caps']:
    cap.set(color ='#8B008B', linewidth = 2)

# changing color and linewidth of medians
for median in bp['medians']:
    median.set(color ='red', linewidth = 3)

# changing style of fliers
for flier in bp['fliers']:
    flier.set(marker ='D', color ='#e7298a', alpha = 0.5)

# x-axis labels
ax.set_xticklabels(['data_1', 'data_2', 'data_3', 'data_4'])

plt.title("Customized box plot")
plt.show()
````

# Pandas plot() method

* Pandas đi kèm với một vài chức năng vẽ áp dụng cho các đối tượng DataFrame hoặc series sử dụng thư viện Matplotlib bên trong, điều này có nghĩa là **bất kỳ biểu đồ nào được tạo bởi thư viện Pandas đều là một đối tượng của Matplotlib.**

* Phương thức plot() của Pandas cung cấp một tập hợp các kiểu biểu đồ thông qua đối số từ khóa **kind** để tạo ra các biểu đồ có vẻ ngoài tốt. Giá trị mặc định của đối số kind là **line**.

* **.plot** cũng là một **thuộc tính của các đối tượng DataFrame và series của Pandas**, cung cấp một tập hợp nhỏ các biểu đồ có sẵn với Matplotlib. Trong thực tế, Pandas làm cho việc vẽ biểu đồ đơn giản chỉ bằng cách viết một dòng mã bằng cách tự động hóa phần lớn của quy trình trực quan hóa dữ liệu cho chúng ta.




````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

dataset_url = ('https://raw.githubusercontent.com/m-mehdi/pandas_tutorials/main/weekly_stocks.csv')
df = pd.read_csv(dataset_url, parse_dates=['Date'], index_col='Date')
pd.set_option('display.max.columns', None)
print(df.head())
````

## Line chart



- Biểu đồ mặc định là biểu đồ line, hiển thị chỉ mục trên trục x và các cột số khác trong DataFrame trên trục y.

- Hãy vẽ một biểu đồ line và xem Microsoft hoạt động trong 12 tháng trước:

````python
df.plot(y='MSFT', figsize=(9,6))
````

````python
df.plot.line(y=['FB', 'AAPL', 'MSFT'], figsize=(10,6))
````

````python
df.plot(y='FB', figsize=(10,6), title='Facebook Stock', ylabel='USD')
````

````python
df.plot(y='FB', figsize=(10,6), title='Facebook Stock', ylabel='USD', legend=False)
````

## Bar chart

- Chúng ta sẽ tạo một biểu đồ cột dựa trên **giá cổ phiếu trung bình hàng tháng để so sánh giá trị trung bình của mỗi công ty với các công ty khác trong một tháng cụ thể**.

- Để làm điều này, trước tiên, chúng ta cần resample dữ liệu theo cuối tháng và sau đó sử dụng phương thức mean() để tính toán giá cổ phiếu trung bình trong mỗi tháng. Chúng ta cũng chọn ba tháng cuối cùng của dữ liệu, như sau:



````python
df_3Months = df.resample(rule='M').mean()[-3:]
print(df_3Months)
````

````python
df_3Months.plot(kind='bar', figsize=(10,6), ylabel='Price')
````

````python
df_3Months.plot(kind='barh', figsize=(9,6))
````



- Chúng ta cũng có thể vẽ dữ liệu trên biểu đồ cột dọc (stacked vertical) hoặc cột ngang (horizontal bar), biểu thị các nhóm khác nhau lên trên nhau.

- Chiều cao của cột kết quả cho thấy kết quả kết hợp của các nhóm. Để tạo một biểu đồ cột chồng, chúng ta cần gán True cho đối số **stacked**, như sau:

````python
df_3Months.plot(kind='bar', stacked=True, figsize=(9,6))
````

## Histogram


- Một biểu đồ histogram là một loại biểu đồ cột thể hiện phân phối của dữ liệu số liệu, trong đó trục x biểu thị các khoảng bin trong khi trục y biểu thị tần suất dữ liệu trong một khoảng cụ thể.

````python
## In the example above, the bins argument specifies the number of bin intervals,
##  and the alpha argument specifies the degree of transparency.

df['MSFT'].plot(kind='hist', bins=25, alpha=0.6, figsize=(9,6))
````

````python
df[['MSFT', 'FB']].plot(kind='hist', bins=25, alpha=0.6, figsize=(9,6))
````

````python
## A histogram can also be stacked. Let’s try it out:

df[['MSFT', 'FB']].plot(kind='hist', bins=25, alpha=0.6, stacked=True, figsize=(9,6))
````

## Scatter charts

- Biểu đồ phân tán (scatter plots) biểu thị các điểm dữ liệu trên các trục x và y để hiển thị mối tương quan giữa hai biến.

````python
df.plot(kind='scatter', x='MSFT', y='AAPL', figsize=(9,6), color='Green')
````

**Ta có thể thấy xu hướng là nếu giá stock của Apple tăng thì của Microsoft cũng tăng theo.**

## Hexbin Plot


- Khi dữ liệu rất dày đặc, chúng ta có thể sử dụng biểu đồ bin hexagon (hexbin plot) như một phương thức thay thế cho biểu đồ phân tán.

- Nói cách khác, khi số lượng điểm dữ liệu rất lớn và mỗi điểm dữ liệu không thể được biểu thị riêng lẻ, việc sử dụng loại biểu đồ này, biểu diễn dữ liệu dưới dạng một mảnh tổ ong, là tốt hơn. Ngoài ra, màu sắc của mỗi ô hexbin xác định mật độ của các điểm dữ liệu trong phạm vi đó.

````python
df.plot(kind='hexbin', x='MSFT', y='AAPL', gridsize=20, figsize=(10,6))
````


**Đối số gridsize xác định số lượng hexagon trong hướng x. Một kích thước lưới lớn hơn đồng nghĩa với việc có nhiều và nhỏ hơn các bin. Giá trị mặc định của đối số gridsize là 100.**

## Pie chart


- Nếu chúng ta quan tâm đến tỉ lệ, biểu đồ pie là một biểu diễn tỷ lệ tuyệt vời của dữ liệu số học trong một cột. Ví dụ sau cho thấy phân phối giá cổ phiếu trung bình của Apple trong ba tháng trước:

````python
df_3Months.index=['March', 'April', 'May']
df_3Months.plot(kind='pie', y='AAPL', legend=False, autopct='%1.2f')
````



- Để biểu diễn dữ liệu của tất cả các cột trong nhiều biểu đồ pie như các subplot, chúng ta có thể gán True cho đối số subplots, như sau:

````python
df_3Months.plot(kind='pie', legend=False, autopct='%.f', subplots=True, figsize=(14,8))
````

## Box plot


- Biểu đồ hộp (box plot) bao gồm ba phân vị (quartiles) và hai dải chắn (whiskers) mô tả dữ liệu trong một tập hợp các chỉ số: giá trị minimum, first quartile, median, third quartile, và maximum.

- Một biểu đồ hộp truyền đạt thông tin hữu ích, như phạm vi liên phân vị (interquartile range - IQR), median và các outliers của mỗi nhóm dữ liệu.



````python
df.plot(kind='box', figsize=(9,6))
````


**Chúng ta có thể tạo ra biểu đồ box ngang, giống như biểu đồ thanh ngang, bằng cách gán False cho đối số vert. Như sau:**


````python
df.plot(kind='box', vert=False, figsize=(9,6))
````

## Area chart

- Biểu đồ area là một sự mở rộng của biểu đồ đường (line chart) mà lấp đầy khu vực giữa biểu đồ line và trục x bằng một màu sắc. Nếu có nhiều hơn một biểu đồ area hiển thị trong cùng một biểu đồ, các màu sắc khác nhau phân biệt các biểu đồ area khác nhau.

````python
df.plot(kind='area', figsize=(9,6))
````


Phương thức plot() của Pandas mặc định tạo ra một biểu đồ **stacked area**. Việc bỏ chồng biểu đồ diện tích thông thường được thực hiện bằng cách gán False cho đối số stacked:

````python
df.plot(kind='area', stacked=False, figsize=(9,6))
````

## KDE plot


- Kernel Density Estimate (KDE) trực quan hóa mật độ xác suất của một biến dữ liệu liên tục và phi tham số. Biểu đồ này sử dụng các Gaussian kernel để ước lượng hàm mật độ xác suất (PDF) bên trong.



````python
df.plot(kind='kde')
````

**Chúng ta cũng có thể chỉ định bandwidth ảnh hưởng đến độ mịn của biểu đồ trong KDE plot, như sau:**

````python
df.plot(kind='kde', bw_method=0.1)
````

````python
df.plot(kind='kde', bw_method=1)
````


**Như chúng ta có thể thấy, việc chọn một bandwidth nhỏ dẫn đến việc làm mịn không đủ, điều này có nghĩa là biểu đồ density xuất hiện như một kết hợp của các peaks riêng lẻ. Ngược lại, một bandwidth lớn dẫn đến việc làm mịn quá mức, điều này có nghĩa là biểu đồ density xuất hiện như một phân phối unimodal.**
